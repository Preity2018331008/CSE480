## Exercise 3: Exploring Entanglement
The goals of this exercise are to: 
1. create entanglement
2. run a quantum circuit on a simulator

2. Entangle the qubits such that the circuit, when measured, will either result in the state 01 or 10 with 50% probability of obtaining either state. This means when one qubit is measured to be a 0, the other qubit will be measured to be a 1, always. Hint: the state vector of the desired state is C(|01> + |10>). You will need to apply quantum gates to reach your desired state, and at least one of them will be a two-qubit gate.

3. Apply measurement operators to your quantum circuit.

4. Collect the counts dictionary from the results and store this dictionary in counts.

```python
# Import necessary packages
from qiskit.quantum_info import Statevector
from qiskit.tools.visualization import plot_histogram
from qiskit import QuantumCircuit, qpy, execute
from qiskit_aer import Aer
```

```python
# Create quantum circuit with two qubits
qc = QuantumCircuit(2, 2)

# Apply the Hadamard gate to the first qubit
qc.x(1)
qc.h(0)
qc.cx(0, 1)

# Apply measurement operators to the quantum circuit
qc.measure([0,1], [0,1])
```

```python
# Set backend to Aer Simulator
backend = Aer.get_backend('aer_simulator')

# Run the circuit and get counts
result = backend.run(qc, shots = 1024).result()
counts = result.get_counts()

# Print and plot results
print("Counts:", counts)
plot_histogram(counts)
```
